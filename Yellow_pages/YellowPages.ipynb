```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup
```

```python
r1=requests.get('https://www.paginasamarillas.com.co/servicios/hoteles')
soup=BeautifulSoup(r1.text,'html.parser')
```

```python
# Nombre del hotel

Hotel_Name=soup.findAll(class_='webLink')
Hotel_Name[0]['title'].split('web de')[1] # Nombre hotel
```

```python
# Website del hotel 

Web=Hotel_Name[0]['href']  # url
```

```python
# direccion del hotel 

Direccion=soup.findAll('span',attrs={'class':'directionFig'})
Direccion[0].text.split('\r\n\t\t\t\t\t\t\t\t\t\t')[1].split('\r\n\t\t\t\t\t\t\t\t\t')[0] # Direccion
```

```python
# Localidad

city=soup.findAll('span',attrs={'class':'hide'})
city[4].text.split('\r\n\t\t\t\t\t\t\t\t\t\t\t')[1] # Dexripcion del hotel
```

```python
city[0].text.split('\r\n\t\t\t\t\t\t\t\t\t\t\t')[1]
```

```python
# Telefono

phone=soup.findAll(class_='phoneFig')
phone[0].text
```

# Información pagina 1

```python
Hoteles1 = [] 
for i in range(20):  
    Nombre=Hotel_Name[i]['title'].split('web de')[1]
    Web=Hotel_Name[i]['href']
    City=city[2*i].text.split('\r\n\t\t\t\t\t\t\t\t\t\t\t')[1]
    Direc=Direccion[i].text.split('\r\n\t\t\t\t\t\t\t\t\t\t')[1].split('\r\n\t\t\t\t\t\t\t\t\t')[0]
    tel=phone[i].text
    Hoteles1.append((Nombre,tel,Web,City,Direc))
```

```python
df1 = pd.DataFrame(Hoteles1, columns=['Nombre','Telefono','Web','city','Direc'])
```

```python
df1.head()
```

# Extraccion paginas restantes desde la 2 a la 9


```python
url=[]  # Vector donde se cargaran la información de todas las url's
for i in range(2,500):
    url.append('https://www.paginasamarillas.com.co/servicios/hoteles?page=' + str(i))
```

```python
Hoteles2 = []  

for i in range(9):
    pages=requests.get(url[i])
    soup=BeautifulSoup(pages.text,'html.parser')
    Hotel_Name=soup.findAll(class_='webLink')
    Direccion=soup.findAll('span',attrs={'class':'directionFig'})
    phone=soup.findAll(class_='phoneFig')
    city=soup.findAll('span',attrs={'class':'hide'})
    
       # Ciclo de extraccion por pagina 
    
    for i in range(20):  
        Nombre=Hotel_Name[i]['title'].split('web de')[1]
        Web=Hotel_Name[i]['href']
        City=city[2*i].text.split('\r\n\t\t\t\t\t\t\t\t\t\t\t')[1]
        tel=phone[i].text
        Direc=Direccion[i].text.split('\r\n\t\t\t\t\t\t\t\t\t\t')[1].split('\r\n\t\t\t\t\t\t\t\t\t')[0]
        Hoteles2.append((Nombre,tel,Web,City,Direc))
```

```python
df2_9 = pd.DataFrame(Hoteles2, columns=['Nombre','Telefono','Web','city','Direc'])
```

```python
Vacantes_final=pd.concat([df1,df2_9],ignore_index=True)
```
